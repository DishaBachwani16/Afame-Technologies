## Data cleansing involves removing unnecessary columns.

```python
data=pd.DataFrame(hr_data[['EmployeeNumber','Department','JobRole','StandardHours','HourlyRate','OverTime','MonthlyRate','PercentSalaryHike','YearsAtCompany','YearsInCurrentRole','YearsSinceLastPromotion','YearsWithCurrManager','TrainingTimesLastYear']])
```

```python
data.shape
```

## Giving the columns new names.

```python
new_col_name=['Employee ID','Department','Job Role','Std Hours','Hourly Rate','Over Time','Monthly Rate','% Salary Hike','Years at Company','Years in Current Role','Years since Last Promotion','Years with Current Manager','Training Time Last Year']
len(new_col_name)
```

```python
data.columns=new_col_name
```

```python
data.head()
```

## Eliminating redundant entries.

```python
data.drop_duplicates(subset=['Employee ID'])
```

## Eliminate the dataset's NaN values.

```python
data.dropna()
```

## Visualization

```python
hr_data['JobRole'].unique()
```

```python
hr_data['JobRole'].unique().size
```

```python
job_num=data['Job Role'].value_counts()
job_num
```

```python
plt.figure(figsize=(15,5))
for i in job_num.items():
    p = plt.bar(job_num.index, job_num.values)
    plt.bar_label(p, label_type='center')
plt.xlabel('Job Role')
plt.ylabel('Employee Count')
plt.title('Number of Employees in each Job role')
plt.xticks(rotation='vertical')
plt.show()
```
